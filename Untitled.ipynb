```python
#multimedia
from PIL import Image
from matplotlib.pyplot import imshow
import numpy as np
import os
import csv
%matplotlib inline
```

```python
#color histogram
DATASET = "dataset"
file_data = []
color_histogram_data = list()
for filename in os.listdir(DATASET):
    img = Image.open(DATASET+'/'+filename)
    img_row = [filename,','.join(str(i) for i in img.histogram())]
    color_histogram_data.append(img_row)

    #print img.histogram()
    #print "Loading: %s" % filename
    
f = open("color_histogram.csv","w")
w = csv.writer(f)
w.writerows(color_histogram_data)
f.close()
print 'finish color histogram'
```

```python
#euclidean
def euclidean(arr1,arr2):
    if len(arr1) != len(arr2):
        return None
    sum = 0
    for i in xrange(len(arr1)):
        sum += (arr1[i]-arr2[i])**2
    return sum**0.5
```

```python
#match
def match(img):
    match = list()
    img_histogram = img.histogram()
    print img,img_histogram
    f = open('color_histogram.csv','r')
    
    for row in csv.reader(f):
        distance = euclidean(img_histogram,[int(i) for i in row[1].split(',')])
        if distance == None:
            continue
        match_row = (row[0],distance)
        match.append(match_row)
    f.close()
    
    match = sorted(match,key=lambda match:match[1])
    print match[0:5]
    return match[0][0]
```

```python
#Photo Mosaic
def mosaic(img_name,n):
    img = Image.open(img_name)
    width,height=img.size
    bw = width/n
    bh = height/n
    
    for w in xrange(0,width,bw):
        for h in xrange(0,height,bh):
            box =(w,h,w+bw,h+bh)
            b_img = img.crop(box)
            m_img = match(b_img)
            print m_img,box
            m_img = Image.open('dataset/'+m_img)
            m_img = m_img.resize((bw,bh))
            img.paste(m_img,box)
    return img
```
